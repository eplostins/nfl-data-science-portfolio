```python
import pandas as pd

# Load the uploaded CSV
df = pd.read_csv('passing_summary-2.csv')

# Show the first few rows
df.head()
```

```python
# See number of rows and columns
print("Shape:", df.shape)

# View column names
print("Columns:", df.columns.tolist())

# Check for missing data
print("Missing values:\n", df.isnull().sum())

# Summary statistics (mean, std, etc.)
df.describe()

```

```python
y = df['grades_pass']  # PFF Passing Grade (Passing)
features = [
    'accuracy_percent',
    'ypa',
    'avg_depth_of_target',
    'btt_rate',
    'twp_rate',
    'drop_rate',
    'avg_time_to_throw',  # Only weakly correlated with others
    'def_gen_pressures' # is fine to keep if you drop 'sacks'
]

data = df[features + ['grades_pass']].dropna()
X = data[features]
y = data['grades_pass']

```

```python
import statsmodels.api as sm

# Add a constant (intercept term) to the model
X_const = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X_const).fit()

# Show the full regression results
print(model.summary())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
# Risk vs reward
df['btt_to_twp'] = df['btt_rate'] / (df['twp_rate'] + 1e-5)

# Accuracy scaled by production
df['accuracy_ypa'] = df['accuracy_percent'] * df['ypa']

# Risk-adjusted gain
df['net_gain_per_throw'] = df['ypa'] - (df['twp_rate'] * 10)

# Downfield aggression
df['adjusted_btt'] = df['btt_rate'] * df['avg_depth_of_target']

# Sack efficiency
df['pressure_to_sack_ratio'] = df['sacks'] / (df['def_gen_pressures'] + 1)

# Time under pressure
df['ttt_vs_pressure'] = df['avg_time_to_throw'] * df['def_gen_pressures']

df['slow_thrower'] = (df['avg_time_to_throw'] > 2.75).astype(int)
```

```python
engineered_features = [
    'accuracy_percent',
    'ypa',
    'avg_depth_of_target',
    'btt_rate',
    'twp_rate',
    'drop_rate',
    'avg_time_to_throw',
    'def_gen_pressures',

    # Engineered
    'btt_to_twp',
    'accuracy_ypa',
    'net_gain_per_throw',
    'adjusted_btt',
    'pressure_to_sack_ratio',
    'ttt_vs_pressure'
]
```

```python
import statsmodels.api as sm

# Drop rows with missing values
data = df[engineered_features + ['grades_pass']].dropna()
X_new = data[engineered_features]
y_new = data['grades_pass']

# Add constant and fit model
X_new_const = sm.add_constant(X_new)
model_new = sm.OLS(y_new, X_new_const).fit()

# Show regression summary
print(model_new.summary())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Add engineered features
df['btt_to_twp'] = df['btt_rate'] / (df['twp_rate'] + 1e-5)
df['net_gain_per_throw'] = df['ypa'] - (df['twp_rate'] * 10)
df['ttt_vs_pressure'] = df['avg_time_to_throw'] * df['def_gen_pressures']

# 2. Select clustering features
cluster_features = [
    'accuracy_percent',
    'ypa',
    'avg_depth_of_target',
    'btt_rate',
    'twp_rate',
    'drop_rate',
    'avg_time_to_throw',
    'def_gen_pressures',
    'btt_to_twp',
    'net_gain_per_throw',
    'ttt_vs_pressure'
]

# 3. Filter QBs with enough data
df_cluster = df[df['dropbacks'] > 150].copy()

# 4. Prepare data matrix
X_cluster = df_cluster[cluster_features].dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# 5. Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster.loc[X_cluster.index, 'Cluster'] = kmeans.fit_predict(X_scaled)

# 6. Assign Play Style Labels (manual interpretation after viewing cluster means)
# Final cluster-to-label mapping based on analysis
cluster_map = {
    0: "Aggressive Gunslinger",
    1: "Efficient Short-Passer",
    2: "Under Pressure Risk-Taker"
}

# Apply the mapping
df_cluster['Play_Style'] = df_cluster['Cluster'].map(cluster_map)

df_cluster['Play_Style'] = df_cluster['Cluster'].map(cluster_map)

# 7. PCA for 2D Visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df_cluster.loc[X_cluster.index, 'PC1'] = X_pca[:, 0]
df_cluster.loc[X_cluster.index, 'PC2'] = X_pca[:, 1]

# 8. Visualize the clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df_cluster, x='PC1', y='PC2', hue='Play_Style', palette='Set2', s=100)

# Label each QB
for _, row in df_cluster.loc[X_cluster.index].iterrows():
    plt.text(row['PC1'] + 0.1, row['PC2'], row['player'], fontsize=8)

plt.title("QB Clusters by Play Style (PCA Projection)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.tight_layout()
plt.show()

# 9. Optional: Export labeled QB profiles
df_cluster[['player', 'team_name', 'Play_Style', 'Cluster'] + cluster_features].to_csv("qb_clusters.csv", index=False)
```
